```python
import json
```

```python
with open('_news/news.json', 'r') as f:
    news = json.load(f)
```

```python
news = [{**n, 'selected': False} for n in news]
```

```python
news[0]
```

```python
for n in news:
    if 'assets/images/blog/' in n['image']:
        n['image'] = n['image'].replace('assets/images/blog/', 'assets/img/news/')
```

```python
news[0]
```

```python
for n in news:
    if ('http://'in n['link']) or ('https://'in n['link']) :
        n['inline'] = True
    else:
        n['inline'] = False
```

```python
news
```

```python
for n in news:
    n['categories'] = n['category']
    del n['category']
```

```python
# revert order of news and assign progressive id
news = list(reversed(news))
for i, n in enumerate(news):
    n['id'] = i
news
```

```python
for n in news:
    n['date_verb'] = n['date']
    del n['date']
```

```python
news
```

```python
# save news to xai_news.json
with open('_news/xai_news.json', 'w') as f:
    json.dump(news, f)
```

```python
news_date_verb=[{'id':n['id'], 'date_verb':n['date_verb']} for n in news]
```

```python
with open('_news/news_date_verb.json', 'w') as f:
    json.dump(news_date_verb, f)
```

```python
with open('_news/xai_news_date.json', 'r') as f:
    allDates = json.load(f)
```

```python
# join allDates with news on id
for n in news:
    n['date'] = allDates[n['id']]['date']
news
```

```python
with open('_data/xai_news_all.json', 'w') as f:
    json.dump(news, f)
```
